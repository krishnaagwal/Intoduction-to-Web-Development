# Introduction to web development




 1. Explain the main components of client-server architecture and their functions?
   - Main Components of Client-Server Architecture and Their Functions

- Client: Requests data/services from the server (e.g., browsers, mobile apps).

- Server: Processes requests and sends responses (e.g., web servers, databases).

- Database: Stores and manages data for the server to access.

- Network: Facilitates communication between client and server (e.g., HTTP, WebSocket).

2. Discuss the various career paths available after completing a web development course?

 - Various Career Paths After Completing a Web Development Course

- Frontend Developer: Builds user interfaces using HTML, CSS, - JavaScript, React, etc.
- Backend Developer: Manages server logic, databases, and APIs (Node.js, Express, MongoDB).
- Full-Stack Developer: Works on both frontend and backend.
- UI/UX Designer: Focuses on user experience and interface design.
- DevOps Engineer: Manages deployment, CI/CD, and cloud infrastructure.

4. What are the advantages of using a Version Control System like Git in a development project?
 - The Advantages of using version control are:
- Collaboration: Enables multiple developers to work on the same project.
- Version History: Tracks changes and allows rollbacks.
- Branching & Merging: Facilitates parallel development without conflicts.
- Backup & Security: Ensures code safety and redundancy.
